New types of objects can be defined using **classes**. <br>

Classes can be a bit abstract, but are really useful for solving complex models. You can view them as advanced containers that contain both data (accessed as **attributes**) and functions specific to the class (**methods**).

A common approach to solving models using classes is to define the model as a class: the parameters are stored as attributes, and the operations required to solve the model become the methods.

## 1. <a id='toc1_'></a>[Basic example](#toc0_)

```python
class Human():
    
    def __init__(self,name,height,weight): # called when created (initialize)
        
        print('Creating a new human object called ' + name) # print a message

        # save the inputs as attributes
        self.name = name # an attribute
        self.height = height # an attribute
        self.weight = weight # an attribute
    
    def bmi(self): # a method
        
        bmi = self.weight/(self.height/100)**2 # calculate bmi
        return bmi # output bmi
    
    def print_bmi(self): # another method

        print(self.bmi())
```

A class is used as follows:

```python
# a. create an instance of the human object called "jeppe"        
jeppe = Human('jeppe',182,80) # height=182, weight=80
print(type(jeppe))

# b. print an attribute
print(jeppe.height)

# c. call a method (with print statement)
jeppe.print_bmi()
```

**Methods** are like functions, but can automatically use all the attributes of the class (saved in *self.*) without getting them as arguments.

**Key concept — `self`:** Every method receives `self` as its first argument. `self` refers to the specific instance you are working with. When you write `jeppe.bmi()`, Python automatically passes `jeppe` as `self`, so inside the method `self.weight` means `jeppe.weight`. You never pass `self` explicitly when calling a method.

**Attributes** can be changed and extracted with **.-notation**

```python
jeppe.height = 160
print(jeppe.height)
print(jeppe.bmi())
```

Or with **setattr- and getattr-notation** (set attribute/get attribute)

```python
setattr(jeppe,'height',182) # jeppe.height = 182
height = getattr(jeppe,'height') # height = jeppe.height
print(height)
print(jeppe.bmi())
```

## 2. <a id='toc2_'></a>[Operator methods](#toc0_)

If the **appropriate methods** are defined, standard operators, e.g. +, and general functions such as print can be used.

Define a new type of object called a **Fraction**:

```python
class Fraction:
    
    def __init__(self,numerator,denominator): # called when created
        self.num = numerator
        self.denom = denominator
    
    def __str__(self): # called when using print
        
        return f'{self.num}/{self.denom}' # string = self.num/self.denom
    
    def __add__(self,other): # called when using +
        
        new_num = self.num*other.denom + other.num*self.denom
        new_denom = self.denom*other.denom
        
        return Fraction(new_num,new_denom)
    
```

**Note:** We use that 

$$\frac{a}{b}+\frac{c}{d}=\frac{a \cdot d+c \cdot b}{b \cdot d}$$

We can now **add fractions**:

```python
x = Fraction(1,3)
print(x)
```

```python
x = Fraction(1,3) # 1/3 = 5/15
y = Fraction(2,5) # 2/5 = 6/15
z = x+y # 5/15 + 6/15 = 11/15
print(z,type(z))
```

Equivalent to:

```python
z_alt = x.__add__(y)
print(z,type(z))
```

But we **cannot multiply** fractions (yet):

```python
try:
    z = x*y
    print(z)
except:
    print('multiplication is not defined for the fraction type')
```

**Task:** Implement multiplication for fractions. <br>
Hint: `__mul__` is the name for `*`, like `__add__`is the name for `+`. <br>
But you can also just add a method for multiplication, and call it whatever you want.

**Answer**

```python
class Fraction:
    
    def __init__(self,numerator,denominator): # called when created
        self.num = numerator
        self.denom = denominator
    
    def __str__(self): # called when using print
        
        return f'{self.num}/{self.denom}' # string = self.num/self.denom
    
    def __add__(self,other): # called when using +
        
        new_num = self.num*other.denom + other.num*self.denom
        new_denom = self.denom*other.denom
        
        return Fraction(new_num,new_denom)

    def __mul__(self,other):

        new_num = self.num*other.num
        new_denom = self.denom*other.denom

        return Fraction(new_num,new_denom)
    
```

```python
x = Fraction(1,3) # 1/3 
y = Fraction(2,5) # 2/5 
z = x*y # 1/3*2/5 = 2/15
print(z,type(z))
```

## 3. <a id='toc3_'></a>[Economic model](#toc0_)

Classes can help simplify and structure your code, especially when solving economic models.

Another example is defining an economic agent as a class.

We'll keep things as simple as possible, and look at an agent with a Cobb-Douglas utility function.

$$
u(x_1,x_2)= x_{1}^{\alpha}x_{2}^{1-\alpha}  
$$
With income, $I$, and prices $p_1$ and $p_2$, we know the solution is:
$$
x_{1}^{*} = \alpha \frac{I}{p_1}
$$

$$
x_{2}^{*} = (1-\alpha) \frac{I}{p_2}
$$

```python
class Agent:

    def __init__(self,**kwargs):
        """Initialize agent with default parameters; override any with keyword arguments."""
        
        self.name = 'Asker'
        self.alpha = 0.5
        self.p1 = 1
        self.p2 = 2
        self.I = 10
        
        self.x1 = None
        self.x2 = None
        self.solved = False

        for key, value in kwargs.items():
            setattr(self,key,value) # like self.key = value

    def __str__(self):
        """Return a text summary of the agent's parameters and solution status."""

        # f-strings allow including variables inside {} brackets
        # α is the unicode character for α
        text = f'The Agent is called {self.name} and faces this problem: \n' 
        text += f'Income = {self.I} \n'
        text += f'α = {self.alpha} \n'
        text += f'Prices are (p1,p2) = ({self.p1},{self.p2})\n'
        if self.solved:
            text += f'Optimal consumption of (x1,x2) = ({self.x1},{self.x2})'
        else:
            text += 'Optimal consumption of (x1,x2) has not been found'
        return text

    def u_func(self,x1,x2):
        """Cobb-Douglas utility: x1^alpha * x2^(1-alpha)."""

        return x1**self.alpha*x2**(1-self.alpha)

    def expenditure(self,x1,x2):
        """Total expenditure: p1*x1 + p2*x2."""

        return x1*self.p1+x2*self.p2

    def solve(self):
        """Solve for optimal consumption using the Cobb-Douglas demand formulas."""

        self.x1 = self.alpha * self.I / self.p1
        self.x2 = (1-self.alpha) * self.I / self.p2
        self.solved = True

    def print_solution(self):        
        """Print optimal consumption, utility, and expenditure."""

        text = f'Optimal consumption of (x1,x2) = ({self.x1},{self.x2}) \n'
        text += f'Utility is {self.u_func(self.x1,self.x2):.2f}\n'
        text += f'Expenditure is {self.expenditure(self.x1,self.x2)}'
        
        print(text)
```

```python
Asker = Agent()
print(Asker)
```

```python
Asker.solve()
print(Asker)
```

**This is the pattern you will reuse.** When you build your own models, structure them like `Agent`:

- **Parameters** (`alpha`, `p1`, `p2`, `I`) become **attributes** set in `__init__`.
- **Solving** the model becomes a **method** (`solve`) that stores the result back in attributes (`x1`, `x2`).
- Helper calculations (`u_func`, `expenditure`) and reporting (`print_solution`) become their own methods.

Bundling data and code in one object keeps everything together, so you can create several agents with different parameters and solve them independently.

## 4. <a id='toc4_'></a>[Summary](#toc0_)

The take-aways are:

1. **A class is a user-defined type**
2. **Attributes** are like **variables** encapsulated in the class
3. **Methods** are like **functions** encapsulated in the class — `self` gives them access to all attributes automatically
4. **Operator methods** (`__str__`, `__add__`, ...) let standard Python operators work on your class
5. Methods should have a **docstring** describing what they do
