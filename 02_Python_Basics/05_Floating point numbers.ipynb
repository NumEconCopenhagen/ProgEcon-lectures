## 1. <a id='toc1_'></a>[Formula](#toc0_)

On a computer the real line is approximated with numbers on the form:

$$\text{number} = \text{sign} \times \text{significand} \times \text{base}^{\text{exponent}}$$

For a standard 64-bit number the `base` is 2 and the bits are allocated as:
* **sign**: 1 bit (positive or negative)
* **exponent**: 11 bits
* **significand**: 52 bits

Obviously, this is a finite approximation. 


## 2. <a id='toc2_'></a>[Examples](#toc0_)

All numbers can therefore ***not*** be represented exactly. A *close* neighboring number is thus used.

```python
x = 0.1
print(f'{x:.50f}') # printing x with 50 decimals
x = 17.2
print(f'{x:.50f}') # printing x with 50 decimals
```

Arithmetic on these approximate numbers inherits the small errors. The classic example:

```python
print(0.1 + 0.2)        # not exactly 0.3
print(0.1 + 0.2 == 0.3) # False
```

Simple sums might, consequently, not be exactly what you expect.

```python
d = 0.0
for i in range(10):
    d += 0.1
print(d)
```

And just as surprising:

```python
print(0.1 == 0.10000000000000001)
```

**Comparisons of floating point numbers** is therefore always problematic.<br>
We know that 

$$\frac{a \cdot c}{b \cdot c} = \frac{a}{b}$$

but:

```python
a = 100
b = 0.3
c = 10
equality = ((a*c)/(b*c) == a/b)
print('Does equality hold?', equality)
```

**The fix** is to compare with a *tolerance* instead of exact equality. The smallest distinguishable relative gap is the *machine epsilon* ($\approx 2.2 \times 10^{-16}$), and `math.isclose` uses a tolerance based on it:

```python
import sys
print(sys.float_info.epsilon) # machine epsilon: smallest relative gap

import math
print(math.isclose((a*c)/(b*c), a/b)) # True - equal within a tolerance
```

**Underflow**: Multiplying many small numbers can result in an exact zero:

```python
x = 1e-60
y = 1
for _ in range(6):
    y *= x
    print(y)
```

**Overflow**: If intermediate results are too large to be represented, the final result may be wrong or not possible to calculate:

```python
x = 1.0
y = 2.7
for i in range(200):    
    x *= (i+1)
    y *= (i+1) 
print(y/x) # should be 2.7
print(x,y)
```

**Note:** `nan` is not-a-number. `inf` is infinite.

Here the result is `nan` even though mathematically $y/x = 2.7$: the intermediate value $200!$ is too large to represent, so both `x` and `y` overflow to `inf`, and `inf/inf` is undefined (`nan`).

**Order of additions matter**, but not by that much:

```python
sum1 = 10001234.0 + 0.12012 + 0.12312 + 1e-5
sum2 = 1e-5 + 0.12312 + 0.12012 + 10001234.0
print(sum1-sum2)
```

## 3. <a id='toc3_'></a>[Summary](#toc0_)

The take-aways are:

1. Decimal numbers are **approximate** on a computer!
2. **Never compare floats with equality** (only use strict inequalities)
3. Underflow and overflow can create problems (not very important in practice)
