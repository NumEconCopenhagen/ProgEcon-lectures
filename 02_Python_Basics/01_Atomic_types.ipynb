## 1. <a id='toc1_'></a>[What is a variable?](#toc0_)

A **variable** in Python is a **reference** to an **object** in memory.

Object identity can be inspected with the built-in `id()` function in a given Python session.

Objects have different **types**. The four basic **atomic types** are:

1. Integers
2. Floats (decimal numbers)
3. Strings
4. Booleans (true/false)

```python
import sys # built-in module to see how much memory objects fill
```

## 2. <a id='toc2_'></a>[Integers (int): $-\infty, \ldots$, -1, 0, 1, $\ldots, \infty$](#toc0_)
Python integers are arbitrary precision in practice (limited by available memory and time).

```python
# variable x references an integer type object with a value of 1
x = 1

print('x is a', type(x)) # prints the type of x
print('x =', x)
print('Memory address of x is',id(x)) 
print('x uses',sys.getsizeof(x),'bytes')
```

```python
# variable y references the same integer and therefore has the same memory address as x
y = 1
print('Memory address of y is',id(y)) 
```

```python
# variable z references a new integer type object with a value of 2
z = 2
print('z =', z)
print('Memory address of z is',id(z)) 
```

## 3. <a id='toc3_'></a>[Decimal numbers (float): 3.14, 2.72, 1.0, etc.](#toc0_)

```python
x = 1.2

# variable x references a floating point (decimal number) type object
# with a value of 1.2

print('x is a',type(x))
print('x =',x)
print('x uses',sys.getsizeof(x),'bytes')
```

## 4. <a id='toc4_'></a>[Strings (str): 'abc', '123', 'this is a full sentence', etc.](#toc0_)

```python
x = 'abc'

# variable x references a string type object
# with a value of 'abc'

print('x is a',type(x))
print('x =',x)
print('x uses',sys.getsizeof(x),'bytes')
```

**Note:** Alternatively, use double quotes instead of single quotes.

```python
x = "abcd" 
# variable x references a string type object 
# with a value of 'abcd'

print('x is a',type(x))
print('x =',x)
print('x uses',sys.getsizeof(x),'bytes')
```

## 5. <a id='toc5_'></a>[Booleans (bool): True and False](#toc0_)

```python
x = True 
# variable x references a boolean type object 
# with a value of True

print('x is a',type(x))
print('x =',x)
print('x uses',sys.getsizeof(x),'bytes')
```

We can now also **check types** with `isinstance` method:

```python
x = False
is_bool = isinstance(x, bool) # returns True if x is a boolean type object
print('x is a boolean:', is_bool)
is_int = isinstance(x, int) # returns True if x is an integer type object
print('x is an integer:', is_int)
```

```python
x = 1.2
is_int = isinstance(x, int) # returns True if x is an integer type object
print('x is an integer:', is_int)
is_float = isinstance(x, float) # returns True if x is a float type object
print('x is a float:', is_float)
```

## 6. <a id='toc6_'></a>[Type conversion](#toc0_)

Objects of one type can (sometimes) be **converted** into another type.  

For atomic types in these examples, conversion creates a new object with a new identity.  

As an example, from float to string:

```python
x = 1.2
# variable x references a floating point (decimal number) type object
# with a value of 1.2

y = str(x) 
# variable y now references a string type object 
# with a value created based on x 

print('x =', x)
print('x is a',type(x))
print('y =', y)
print('y is a',type(y))
```

From float to integer with `int()`: truncates toward zero:

```python
x = 2.9

y = int(x) # variable y now references an integer type object  
print('x =', x)
print('y =', y)
print('y is a',type(y))
```

```python
x = -2.9
print('int(x) =', int(x)) # truncates toward zero
```

**Limitation:** You cannot, for example, convert a string **with letters** to an integer.

```python
try: # try to run this block
    x = int('222a')
    print('can be done')
    print(x)
except Exception as e: # if any error found run this block instead
    print('cannot be done')
    print('Error:', e)
    
```

**Note**: Indentation is required and is typically 4 spaces.

But you can convert a string with only numbers to an integer (or float).

```python
x = float('22.2')
print('x =', x)
y = float('222')
print('y =', y)
```

## 7. <a id='toc7_'></a>[Operators](#toc0_)

Variables can be combined using **arithmetic operators** (e.g. +, -, /, **).<br>For numbers we have:

```python
x = 3
y = 2
print(x+y)
print(x-y)
print(x/y)
print(x*y)
print(x**2)
```

For strings we can use an overloaded '+' for concatenation:

```python
x = 'abc'
y = 'def'
print(x+y)
```

A string can also be multiplied by an integer:

```python
x = 'abc'
y = 2
print(x*y)
```

**Question**: What is the result of `x = 3**2`?

- **A:** `x = 3`
- **B:** `x = 6`
- **C:** `x = 9`
- **D:** `x = 12`
- **E:** Don't know


**Note:** Standard division converts integers to floating point numbers.

```python
x = 8
x_neg = -8

y = x/2      # standard division
z = x//3     # floor division
q = x_neg//3 # floor division with a negative number

print(y, type(y))
print(z, type(z))
print(q, type(q))
```

## 8. <a id='toc8_'></a>[Augmentation](#toc0_)

Variables can be changed using **augmentation operators** (e.g. +=, -=, *=, /=)

```python
x = 3 
print('x =',x)

x += 1 # same result as x = x+1
print('x =',x)
x *= 2 # same result as x = x*2
print('x =',x)
x /= 2 # same result as x = x/2
print('x =',x)
```

## 9. <a id='toc9_'></a>[Logical operators](#toc0_)

Variables can be compared using **boolean operators** (e.g. ==, !=, <, <=, >, >=). 

```python
x = 3
y = 2
z = 10
print(x < y) # less than
print(x <= y) # less than or equal
print(x == y) # equal
print(x != y) # not equal
```

The comparison returns a boolean variable:

```python
z = x < y # z is now a boolean variable
print(z)
```
